Task-01

```python
X = [
 [1, 40],
 [3, 65],
 [2, 45],
 [4, 75],
 [3, 71],
 [3, 90],
 [2, 23],
 [5, 67],
 [1, 10],
 [1, 16]
]

y = [0,1,0,1,1,1,0,1,0,0]
```

Task-02

```python
w= [1,2] # weights for [study_hours, attendance_percentage]
b = 1
lr = 0.01
```

Task-03

```python
def step_activation(inputs, weights, bias):

    z = 0
    for i in range(len(inputs)):
        z += inputs[i] * weights[i]
    z += bias

    if z >= 0:
        return 1
    else:
        return 0
```

Task-04

```python
epochs = 10
for epoch in range(epochs):
    errors = 0
    for i in range(len(X)):
        xi = X[i]
        yi = y[i]

        pred = step_activation(xi, w, b)

        # Calculate error
        error = yi - pred

        # Weight update using learning rate (Î· = lr)
        w[0] += lr * error * xi[0]
        w[1] += lr * error * xi[1]
        b    += lr * error

        if error != 0:
            errors += 1

    print(f"Epoch {epoch+1}, Misclassified samples: {errors}")
    if errors == 0:
        print("Training complete.")
        break
```

Task-05

```python
print("Student Pass/Fail Prediction")
study_hours = float(input("Enter study hours per day: "))
attendance = float(input("Enter attendance percentage: "))

# Use the step_activation function correctly
prediction = step_activation([study_hours, attendance], w, b)

if prediction == 1:
    print(" The student is likely to PASS.")
else:
    print(" The student is likely to FAIL.")
```
